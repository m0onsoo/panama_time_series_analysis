```python
df = pd.read_csv("Gatun_Lake_Water_Level_Monthly_Mean.csv")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generating a sample dataframe
# date_range = pd.date_range(start='2010-01-01', end='2024-05-01', freq='M')


# Creating the lineplot
plt.figure(figsize=(20, 8))
# sns.lineplot(df[df.index >= '2010-01-01'])

df2 = df[df.index >= '2020-01-01']
sns.lineplot(data=df2, x=df2.index, y='GATUN_LAKE_LEVEL(FEET)')
plt.xticks(rotation=45)
plt.gca().xaxis.set_major_locator(plt.MaxNLocator(10))  # Ensure one tick per year approximately
plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%Y'))  # Format x-axis ticks as year

plt.ylim([75, 95])
plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Sample Lineplot with Yearly X-axis Labels Rotated 45 Degrees')
plt.tight_layout()
plt.show()
```
